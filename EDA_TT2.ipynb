##Resumen de variables categóricas

##Visualización de datos

```python
plt.figure(figsize=(1,30))
metodo_counts = data['metodo'].value_counts()
sns.barplot(y=metodo_counts.index, x=metodo_counts.values, orient='h')
plt.title('Distribución de métodos')
plt.xlabel('Método')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()
```
